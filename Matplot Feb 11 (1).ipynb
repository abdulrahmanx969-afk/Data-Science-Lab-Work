```python
import matplotlib.pyplot as plt

x = np.arange(0,3*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x,y_sin)
plt.plot(x,y_cos)
plt.xlabel("X Variable")
plt.ylabel("Trignometric Function")
plt.title("Sin and Cos Function")
plt.show()
```

```python
#Subplot

plt.subplot(2,1,1)
plt.plot(x,y_sin, color = "Purple")
plt.xlabel("X Variable")
plt.ylabel("Trignometric Function")
plt.title("Sin and Cos Function")


plt.subplot(2,1,2)
plt.plot(x,y_cos, color = "Yellow")
plt.xlabel("X Variable")
plt.ylabel("Trignometric Function")
plt.title("Sin and Cos Function")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

p = np.arange(0, 100, 8)

q = p**2
d = np.exp(p)

plt.scatter(p, q)
plt.scatter(p, d/1000)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1, 10, 300)
plt.figure(figsize=(12, 10))

plt.subplot(2, 2, 1)
plt.plot(x, 2 * x + 1, color='blue', linewidth=2)
plt.title('Linear Function', fontsize=14)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, alpha=0.3)

plt.subplot(2, 2, 2)
plt.plot(x, x**2, color='green', linewidth=2)
plt.title('Power Function', fontsize=14)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, alpha=0.3)

plt.subplot(2, 2, 3)
plt.plot(x, np.exp(x), color='red', linewidth=2)
plt.title('Exponential Function', fontsize=14)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, alpha=0.3)

plt.subplot(2, 2, 4)
plt.plot(x, np.log(x), color='purple', linewidth=2)
plt.title('Logarithmic Function', fontsize=14)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
